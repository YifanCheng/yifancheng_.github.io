Rather than just recreating examples from each of these projects, I figured I'd make an honest effort to use some real data and create some plots using each.  So for this post, I'll be using the [R-ArcticNET](http://www.r-arcticnet.sr.unh.edu/v4.0/index.html) streamflow database as an example dataset.

Let's start by getting the data.  A bit of funny business but in the end, we get `Pandas DataFrame`s with all our `observations` and `attributes`.

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
print('matplotlib version %s' % mpl.__version__)
import pandas as pd
print('pandas version %s' % pd.__version__)
import seaborn as sns
print('seaborn version %s' % sns.__version__)
# import ggplot
# print('ggplot version %s' % ggplot.__version__)
# import bokeh as bk
# print('bokeh version %s' % bk.__version__)
# import vincent
# print('vincent version %s' % '0.4')
from mpl_toolkits.basemap import Basemap
print('Basemap version %s' % '1.0.7')

# vincent.core.initialize_notebook()
```

```python
import numpy as np
import datetime

def make_dates(years):
    '''make a monthly timeseries given a list of years'''
    dates = []
    for year in years:
        for month in range(1,13):
            dates.append(datetime.datetime(int(year), month, 1))
    return dates

# open attributes file
attributes = pd.read_csv('http://www.r-arcticnet.sr.unh.edu/v4.0/data/SiteAttributes.txt',
                   engine='python',
                   sep= '\t')

# open observations file
data = pd.read_csv('http://www.r-arcticnet.sr.unh.edu/v4.0/data/Discharge_ms.txt',
                   sep= '\t',
                   low_memory=False)

# Month names in observations file
_r_months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

data_series = {}
for i, point in enumerate(attributes['PointID']):
    name = str(attributes['Name'][i])
        
    inds = np.nonzero(data['Point_ID'] == point)[0]
    flow_table = np.empty((len(inds), len(_r_months)))

    for j, month in enumerate(_r_months):
        flow_table[:,j] = data[month][inds]

    times = make_dates(data['Year'][inds])

    data_series[name] = pd.Series(flow_table.flatten(), index=times)

observations = pd.DataFrame(data_series)
```

Let's start by getting an understanding of where all our observations are at.  We'll make a scatter plot using the lat/lon information in the attributes.  

For this set of plots, we'll try to use these packages:
- basemap
- vincent

##Basemap

```python
# Basemap

my_atts = attributes[['Long', 'Lat', 'DArea_effective', 'CountOfYear']].dropna()

fig = plt.figure(figsize=(7, 6))
# get the data that we'll put in the plot
lons = my_atts['Long'].values
lats = my_atts['Lat'].values
sizes = my_atts['DArea_effective'].values / 2000
cols = my_atts['CountOfYear'].values

# draw map with markers for float locations
m = Basemap(projection='npstere',
            lon_0=-113,
            boundinglat=lats.min(),
            resolution='i')
x, y = m(lons, lats)
m.drawcoastlines()
m.drawmapboundary(fill_color='grey') # fill to edge
m.drawcountries()
m.fillcontinents(color='white',lake_color='grey',zorder=0)

m.scatter(x, y, s=sizes, c=cols, marker="o", cmap='Blues', alpha=0.8, vmin=0, vmax=100)
plt.title('r-ArcticNET Streamflow Guages', fontsize=18)
cbar = m.colorbar(location='right', extend='max')
cbar.set_label('Years of Record')
plt.savefig('r-arctic-net-locations.pdf')
```
